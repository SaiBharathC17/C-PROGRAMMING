```python
import pandas as pd
```

```python
dataset = {"MOGID":[101,102,103],"name":["opry","vjd","ganesh"],"height":[180,178,186]}
```

```python
df = pd.DataFrame(dataset)
```

```python
df
```

```python
# @title name vs height

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['name'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='height', y='name', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
